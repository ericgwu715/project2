```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
country_path = "GlobalTemperaturesData/GlobalLandTemperaturesByCountry.csv"
global_path = "GlobalTemperaturesData/GlobalTemperatures.csv"

country_df = pd.read_csv(country_path)
global_df = pd.read_csv(global_path)
```

# Country_df

```python
country_df = country_df[['dt','AverageTemperature','Country']]
country_df = country_df.rename(columns={
    "dt":"date",
    "AverageTemperature":"avgtempcountry",
    "Country":"country"
})

country_df.head()
```

# North

```python
USA_df = country_df.loc[country_df['country']=='United States']

start_date = "1850-01-01"
end_date = "2016-01-01"

after_start_date = USA_df["date"] >= start_date
before_end_date = USA_df["date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates_USA = USA_df.loc[between_two_dates]

filtered_dates_USA['date'] = pd.to_datetime(filtered_dates_USA['date'])

USA_new = filtered_dates_USA.groupby((filtered_dates_USA['date'].dt.year//5)*5).mean()

USA_new = USA_new.rename(columns={
    'avgtempcountry':'avgtempUSA'
})

USA_new.head()

#USA_new.plot()
```

# Merge Data

```python
two_m = USA_new.merge(germany_new, on='date')
three_m = two_m.merge(sweden_new, on='date')
four_m = three_m.merge(venezuela_new, on='date')
five_m = four_m.merge(congo_new, on='date')
six_m = five_m.merge(indonesia_new, on='date')
seven_m = six_m.merge(argentina_new, on='date')
eight_m = seven_m.merge(new_z_new, on='date')
nine_m = eight_m.merge(south_africa_new, on='date')

nine_m
```

```python
fig = plt.figure(figsize=(12,15))

ax = fig.add_subplot(1,1,1)

nine_m.plot(ax=ax,lw=3,alpha=0.75)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.grid()
```

# Global_df

```python
global_df = global_df.rename(columns={
    "dt":"date",
    "LandAverageTemperature":"avglandtemp",
    "LandAverageTemperatureUncertainty":" ",
    "LandMaxTemperature":"do i care",
    "LandMaxTemperatureUncertainty":" ",
    "LandMinTemperature":"no",
    "LandMinTemperatureUncertainty":" ",
    "LandAndOceanAverageTemperature":"avglandoceantemp",
    "LandAndOceanAverageTemperatureUncertainty":" "
})

global_df = global_df[['date','avglandtemp','avglandoceantemp']]

global_df.head()
```

```python
global_df['date'] = pd.to_datetime(global_df['date'])

start_date = "1850-01-01"
end_date = "2016-01-01"

after_start_date = global_df["date"] >= start_date
before_end_date = global_df["date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = global_df.loc[between_two_dates]

yearly = filtered_dates.groupby(global_df['date'].dt.year).mean()

yearly.head()
```

```python
global_new = filtered_dates.groupby((filtered_dates['date'].dt.year//5)*5).mean()

global_new = global_new.rename(columns={
    'avglandtemp':'avgtempGlobal L',
    'avglandoceantemp':'avgtempGlobal'
})

global_new = global_new[['avgtempGlobal']]

global_new.head()

#df2.to_csv('fiveyr.csv')
```

```python
plt.plot(global_new['avgtempGlobal'])
```

# Ultimate Merge

```python
all_m = nine_m.merge(global_new, on='date')

all_m.head(10)
```

```python
fig = plt.figure(figsize=(12.5,10))

ax = fig.add_subplot(1,1,1)

all_m.plot(ax=ax,lw=3,alpha=0.75)

ax.legend(loc='best', bbox_to_anchor=(1, 0.5))

ax.grid()

plt.savefig('graph/extractedTemps.pdf')
```
